## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Removing duplicates
credits_clean = credits.drop_duplicates()
titles_clean = titles.drop_duplicates()

# Filling missing values for 'character' in credits and titles
credits_clean['character'].fillna('Unknown', inplace=True)
titles_clean['description'].fillna('No description available', inplace=True)
titles_clean['age_certification'].fillna('Not Rated', inplace=True)
titles_clean['seasons'].fillna(0, inplace=True)
titles_clean['imdb_id'].fillna('Unknown', inplace=True)
titles_clean['imdb_score'].fillna(titles_clean['imdb_score'].mean(), inplace=True)
titles_clean['imdb_votes'].fillna(0, inplace=True)
titles_clean['tmdb_popularity'].fillna(titles_clean['tmdb_popularity'].mean(), inplace=True)
titles_clean['tmdb_score'].fillna(titles_clean['tmdb_score'].mean(), inplace=True)

titles_clean['release_year'] = titles_clean['release_year'].astype(int)
titles_clean['seasons'] = titles_clean['seasons'].astype(int)
titles_clean['title'] = titles_clean['title'].str.title()
titles_clean['genres'] = titles_clean['genres'].str.replace(',', ', ').str.title()
titles_clean['decade'] = (titles_clean['release_year'] // 10) * 10


# Converting data types where necessary
titles_clean['release_year'] = titles_clean['release_year'].astype(int)
titles_clean['seasons'] = titles_clean['seasons'].astype(int)

# Standardizing text data
titles_clean['title'] = titles_clean['title'].str.title()
titles_clean['genres'] = titles_clean['genres'].str.replace(',', ', ').str.title()

# Creating new features for analysis
titles_clean['decade'] = (titles_clean['release_year'] // 10) * 10

# Verifying if missing values are handled
print('Cleaned Credits Missing Values:\n', credits_clean.isnull().sum())
print('Cleaned Titles Missing Values:\n', titles_clean.isnull().sum())

# Final shape of cleaned datasets
print('Cleaned Credits Shape:', credits_clean.shape)
print('Cleaned Titles Shape:', titles_clean.shape)
````

````python
print('Credits Head:\n', credits.head())
print('Titles Head:\n', titles.head())
````

### What all manipulations have you done and insights you found?

The data wrangling process involved several essential steps to prepare the datasets for analysis. First, we removed duplicate rows from both the credits and titles datasets to ensure data integrity. We then addressed missing values by filling them with appropriate placeholders or statistical measures: the character column in the credits dataset was filled with 'Unknown', while columns like description, age_certification, and imdb_id in the titles dataset were also assigned meaningful default values. For numeric columns, such as imdb_score, tmdb_popularity, and tmdb_score, missing values were replaced with their respective column means, ensuring the overall distribution of data remained intact. Additionally, seasons and imdb_votes were set to 0 where data was missing.

To maintain consistency, we converted the release_year and seasons columns to integers and standardized the text data by capitalizing the title column and formatting genres with proper spacing and title case. As part of feature engineering, we created a new decade column by grouping movies and shows based on their release year, providing a clearer perspective on trends over time.

With these transformations, the datasets are now clean, consistent, and well-structured, free from missing values and duplicates. They are ready for in-depth analysis, enabling more accurate insights and visualizations.
